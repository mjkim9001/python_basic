```python
# 한글 폰트 사용(Matplotlib사용 전 기본 설정해주기)
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font', family = 'Malgun Gothic')
```

```python
#1
import csv

f= open('hospital_2019.csv', 'r', encoding = 'utf=8')
lines = csv.reader(f)

header = next(lines)

count=0

for line in lines:
    if line[2]=='전남':
        count+=1
        print('%d.' %count,line[0],'/', line[3],'/', line[6])
f.close()
```

```python
#2
f= open('hospital_2019.csv', 'r', encoding = 'utf=8')
lines = csv.reader(f)

header = next(lines)
count=0

for line in lines:
    if line[2]=='제주' and 2015<int(line[5][:4])<2019 and line[1]=='의원':
        count+=1
        
print('2016~2018년 3개년간 개원한 제주도의 의원 수: %d개' %count)
f.close()
```
