## ch13 연습문제

```python
df = pd.read_csv('high_school_2019.csv', skiprows=[0],
                names=['지역', '학교', '1학년 학습수', '1학년 학생수','2학년 학급수','2학년 학생수',
                       '3학년 학급수','3학년 학생수'])
```

```python
df1 = df.astype({'1학년 학생수':int})
```

```python
df['총 학생수']= df1[['1학년 학생수', '2학년 학생수','3학년 학생수']].sum(axis=1)
df
```

```python
result = df1[['1학년 학생수', '2학년 학생수','3학년 학생수']].sum()
result.index = ['1학년','2학년','3학년']
result
```

```python
result.plot.bar(rot=1)
plt.title('전국 고등학교 학년별 총 학생수')
plt.show()
```
