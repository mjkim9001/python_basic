```python
# Q 12-3
import random

def make_rand(n):
    result = []
    for i in range(n):
        result.append(random.randrange(1,7))
    return result

count = 10
y = make_rand(count)
x = list(range(1,11))
plt.xticks(x)
plt.yticks([1,2,3,4,5,6])

plt.title('주사위 10회 던지기')
plt.bar(x, y)
plt.xlabel('던진 횟수')
plt.ylabel('주사위눈')

plt.show()
```

```python
#스템 플롯 (굵기가 없는 선으로된 )
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title("Stem Plot")
plt.stem(x, np.cos(x), '-.', use_line_collection = True)
plt.show()
```

```python
# Bubble Chart

N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

```python
# 등고선

def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

```python
# 3D 서피스 플롯
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```
