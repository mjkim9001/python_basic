```python
# ex 12-3 한글 폰트 사용
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font', family = 'Malgun Gothic')
#from matplotlib import rc

#rc('font', family='Malgun Gothic')

font1 = {'size':18, 'color':'green'}

xdata = ['안지영', '홍지수', '황예린']
ydata = [90, 85, 88]
plt.plot(xdata, ydata)

plt.title('세명 학생의 국어 성적', fontdict = font1)
plt.xlabel('이름')
plt.ylabel('성적')

plt.show()
```

```python
# ex12-4 선 그래프의 범례

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
plt.plot(xdata, ydata1, label ='국어')
plt.plot(xdata, ydata2, label = '영어')
plt.legend(loc ='upper center')

plt.title('세명 학생의 국어, 영어, 성적')

plt.show()
```

```python
# ex12-5 그래프 선 스타일 설정

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, label = '국어', color = 'red', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label = '영어', color = '#00ffff', linestyle='--', marker='x')
plt.plot(xdata, ydata3, label = '수학', color = 'magenta', linestyle='-.', marker='s')
plt.plot(xdata, ydata4, label = '사회', color = '#444444', linestyle=':', marker='d')

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
# 간단하게 그릴수 있는 방법

plt.plot(xdata, ydata1, 'ro-', label = '국어', linewidth = 3, markersize = 8)
plt.plot(xdata, ydata2, 'cx--', label = '영어', linewidth = 3, markersize = 8)
plt.plot(xdata, ydata3, 'ms-.', label = '수학', linewidth = 3, markersize = 8)
plt.plot(xdata, ydata4, 'kd:',label = '사회', linewidth = 3, markersize = 8)

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
# ex12-6
xdata = list(range(30))
ydata = []

for x in xdata:
    y = 2 * x
    ydata.append(y)
    
plt.plot(xdata, ydata, color='blue', label= 'y=2x', linewidth = 5)
plt.xlim(0, 30)
plt.ylim(0, 60)
plt.xticks(list(range(-1,30,2)))
plt.yticks(list(range(-2,60,5)))
plt.legend(loc='best')

plt.grid(True)
plt.show()
```

```python
# Sin & Cos

#from math import sin, cos, pi
import math

x = [0.1*i for i in range(0, 21)]

y_sin = [math.sin(j*math.pi) for j in x]
y_cos = [math.cos(j*math.pi) for j in x]

plt.plot(x, y_sin)
plt.plot(x, y_cos, color = 'red')
plt.show()
```

```python
import numpy as np
xs = np.arange(0, 2*np.pi+0.01, 0.01)
ysin = np.sin(xs)
ycos = np.cos(xs)

plt.plot(xs, ysin, label = 'Sine curve')
plt.plot(xs, ycos, label = 'Cosine curve')
plt.title('삼각함수 그래프')
plt.legend(loc = 'best')
plt.xticks(np.arange(0, 2*np.pi+0.01, np.pi/2))


plt.axhline(y=0, color='k', linewidth=0.5) #수평선(x축)
plt.axhline(y=1, color='k', linewidth=0.5)
plt.axhline(y=-1, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5) # 수직선(y축)
plt.axvline(x=np.pi, color='k', linewidth=0.5)
plt.axvline(x=2*np.pi, color='k', linewidth=0.5)
```

```python
# Sigmoid 함수
xs = np.linspace(-3, 3, 601)
ys = 1 / (1+np.exp(-xs))

plt.plot(xs, ys)
plt.ylim(-0.1, 1.1)
plt.title('시그모이드 그래프')

plt.axhline(y=0, color='k', linewidth=0.5)
plt.axhline(y=1, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

plt.show()
```
