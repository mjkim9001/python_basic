```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
from urllib.parse import urljoin
```

```python
url= "https://www.siksinhot.com/taste?upHpAreaId=17&hpAreaId=660"

html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
```

```python
re_list = soup.find_all(class_='cont')
re_list
```

```python
store =[]
menus = [] 
scores = []
url_add = []
n=0
com = ','
for result in re_list[:21]:
    store.append(result.find(class_="store").get_text())
    tmp_str=result.find('p').get_text()
    menus.append(tmp_str.split(' ')[0])
    if menus[n][-1] == com:
        menus[n] = menus[n].replace(',','')
    #scores.append(result.find(class_='score').get_text())
    if result.find(class_='score2')!= None:
        scores.append(result.find(class_='score2').get_text())
    else:
        scores.append(result.find(class_='score').get_text())
    url_add.append(urljoin(url, result.find('a')['href']))
    n+=1
```

```python
print(store)
print(len(store),len(menus),len(scores))
print(menus)
print(scores)
print(url_add)
```

```python
df = pd.DataFrame({'식당명':store, '메인메뉴':menus,'평점':scores})
df
```

```python
addr = []
tel = []
for n in tqdm_notebook(df.index):
    html = urlopen(url_add[n])
    soup_tmp = BeautifulSoup(html, 'html.parser')
    store_info = soup_tmp.find(class_='store_info')
    if store_info.find(class_='txt_adr') ==None:
        addr.append(store_info.find(class_='txt_adr').get_text())
    else:
        addr.append(store_info.find('span').get_text())
    tel.append(store_info.find(class_='p_tel').get_text())
    
```
